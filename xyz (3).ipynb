```python
#List of Outliers index
```

```python
def outliers(df, ft):
    Q1= df[ft].quantile(0.25)
    Q3= df[ft].quantile(0.75)
    IQR= Q3-Q1
    
    lower_bound= Q1- 1.5*IQR
    upper_bound= Q3+ 1.5*IQR
    
    ls =df.index[(df[ft]<lower_bound) | (df[ft]>upper_bound)]
    return ls
```

```python
#Store output indices from column in a list
```

```python
index_list = []

for feature in ['KM Travelled', 'Price Charged', 'Cost of Trip', 'Age', 'Income (USD/Month)', 'Company_Pink Cab', 'Company_Yellow Cab']:
    index_list.extend(outliers(data, feature))
```

```python
index_list
```

```python
#Remove index_list
```

```python
def remove(df, ls):
    ls = sorted(set(ls))
    df = df.drop(ls)
    return df
```

```python
data_cleaned = remove(data, index_list)
```
